## Description of the Data

As seen from the table above, there are 57 rows and 14 columns. 

The "Year" column represents the year food had those particular inflation rates in that row. For example, Under the year column, the first number, 1968.0, represents the year 1968, and all the integers in that row are the percentage increase in food prices per month during the year 1968. 

The middle 12 columns represent the months.

The last column, labeled "Ave" represents the average percentage increase during the entire year. For example, in the first row, we can see that during the year 1968, food prices increased by an average of 3.4%.

The data covers the years 1968-2024. (I did not include 2025, as the year is not over yet so they do not have all the data for 2025)

## Data Structure

(I will be using the "Ave" column, to help me to find the overall shape of the data throughout the years)

First, I am going to make a boxplot of the average price percent increase per year. I am using a boxplot because it will help me to find out the median, the max, and the minimum percent increase over all the years.

```python
sns.boxplot(x=rates["Ave"])
plt.title("Average Percent Increase of Food Prices from 1968-2024")
plt.xlabel("Average Percent Increase")
```

As seen from the boxplot above, the maximum percent increase in food prices in a year from the years 1968-2024 is around 14 (13.7). The minimum percent increase in food prices from the years 1968-2024 is a little over 0 (0.3). 

The median is around 3. There also seems to be a few outliers as seen from the dots towards the right of the boxplot.

Next, I am going to make a histogram to help me find the mean, mode, and help me to see the shape of the data.

```python
plt.hist(rates["Ave"])
plt.title("Average Percent Increase of Food Prices from 1968-2024")
plt.xlabel("Percent Increase")
plt.ylabel("Frequency")
plt.show()
```

As seen from the histogram above, the data seems to be skewed to the right. 

The mode seems to be around 2. The mean is approximately 3.83.

**Work done to calculate mean:**
(took the percent increase amount and multiplied it by frequency)

<br> 9 x 1.5 = 13.5

(added those number together)

I made a scatterplot to attempt to find the correlation

```python
sns.scatterplot(x="Ave", y="Year", data=rates)
plt.title("Average Percent Increase of Food Prices over the years (1968-2024)")
plt.xlabel("Average Percent Increase")
```

```python
import numpy as np
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
x = rates["Ave"].to_numpy()
y = rates["Year"].to_numpy()

spearman_rho, spearman_p = stats.spearmanr(x,y)
print(f"{spearman_rho:.3f}")
```

I used Spearman's since the data is non linear, but even then there seems to be no correlation between the year and the average percent price increase of food per year.
